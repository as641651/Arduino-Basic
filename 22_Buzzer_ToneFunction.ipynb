## Passive buzzer

> 1) They are used to make sounds (like beep, alarm, etc). They come in two variations

>> a) **Two pin** : One digital and the other ground.

>> b) **Three pin**: An additional pin for vcc. Since this version can draw power from arduino, they are more louder.

> 2) To control volume, use PWM digital pin or use a potentiometer

> 3) Use **[tone(PIN, FREQUENCY, DURATION=optional)](https://www.arduino.cc/reference/en/language/functions/advanced-io/tone/)** function to play sound. If **duration** is **not set**, you can stop the sound with **noTone()** function.



### Sketch

<div style="border:2px solid;margin:0 20px 0 20px" ></div> 

```c++
const int buzzer = 8; 

void setup(){ 
  pinMode(buzzer, OUTPUT); // Set buzzer - pin 8 as an output
}

void loop(){
  tone(buzzer, 1000); // Send 1KHz sound signal...
  delay(1000);        // ...for 1 sec
  noTone(buzzer);     // Stop sound...
  delay(1000);        // ...for 1sec
}
```
